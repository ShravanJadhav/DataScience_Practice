```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/dataset 1.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(2,2,1)
sns.distplot(df['age'], color='r').set_title('Age')

ax = fig.add_subplot(2,2,2)
sns.distplot(df['bmi'], color='black').set_title('BMI')

ax = fig.add_subplot(2,2,3)
sns.distplot(df['charges in INR'],color='g').set_title('Charges in INR')

ax = fig.add_subplot(2,2,4)
sns.countplot(df['sex'],color='b').set_title('SEX')

plt.show()
```
